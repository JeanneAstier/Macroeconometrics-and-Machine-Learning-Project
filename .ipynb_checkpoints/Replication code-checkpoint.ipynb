```python
# import data from FRED-MD, vintage of January 2016
database = pd.read_csv(r'C:\Users\jeann\Downloads\2016-02.csv', sep=',')

# set correctly the dates
database = database.drop([0])
database.sasdate=pd.to_datetime(database.sasdate)
database = database.rename(columns={"sasdate": "Date"})
database = database.set_index("Date",drop=True)

# keep only dates bewteen Jan 1960 and Dec 2015
database = database.drop(database[database.index < '1960-01-01'].index)
database = database.drop(database[database.index > '2015-12-01'].index)

nb_col = database.shape[1]

database.head()
```

# 1. b) stationarization of the variables

"_All variables are transformed to achieve stationarity as described in the supplementary materials._"

From supplementary materials: "_In this section, we present a description of the dataset used in this paper. Tables S.1-S.8
describe the data and the transformations that were applied to each variable._"

```python
# assignment of the columns to their treatment group
t_code_1 = ['NAPMPI', 'CES0600000007', 'AWHMAN', 'NAPMEI', 'NAPM', 'NAPMNOI', 'NAPMSDI', 'NAPMII', 'COMPAPFFx', 'TB3SMFFM', 'TB6SMFFM', 'T1YFFM', 'T5YFFM', 'T10YFFM', 'AAAFFM', 'BAAFFM',  'VXOCLSx', 'NAPMPRI']
t_code_2 = ['CUMFNS', 'HWI', 'HWIURATIO', 'UNRATE', 'UEMPMEAN', 'AWOTMAN', 'ISRATIOx', 'UMCSENTx', 'CONSPI',  'FEDFUNDS', 'CP3Mx', 'TB3MS', 'TB6MS', 'GS1', 'GS5', 'GS10', 'AAA', 'BAA', 'S&P div yield']
t_code_3 = []
t_code_4 = ['HOUST', 'HOUSTNE', 'HOUSTMW', 'HOUSTS', 'HOUSTW', 'PERMIT', 'PERMITNE', 'PERMITMW', 'PERMITS', 'PERMITW']
t_code_5 = ['RETAILx', 'W875RX1','INDPRO', 'IPFPNSS', 'IPFINAL', 'IPCONGD', 'IPDCONGD', 'IPNCONGD', 'IPBUSEQ', 'IPMAT', 'IPDMAT', 'IPNMAT', 'IPMANSICS', 'IPB51222S', 'IPFUELS', 'CLF16OV', 'CE16OV',
           'UEMPLT5', 'UEMP5TO14', 'UEMP15OV', 'UEMP15T26', 'UEMP27OV', 'CLAIMSx', 'PAYEMS', 'USGOOD', 'CES1021000001', 'USCONS', 'MANEMP', 'DMANEMP', 'NDMANEMP', 'SRVPRD', 'USTPU', 'USWTRADE', 'USTRADE', 'USFIRE', 'USGOVT',
           'DPCERA3M086SBEA', 'CMRMTSPLx', 'RETAILx', 'ACOGNO', 'AMDMNOx', 'ANDENOx', 'AMDMUOx', 'BUSINVx', 'M2REAL','TWEXMMTH', 'EXSZUSx', 'EXJPUSx', 'EXUSUKx', 'EXCAUSx', 'S&P 500', 'S&P: indust', 'S&P PE ratio', 'PPIFGS',
            'PPIFCG', 'PPIITM', 'PPICRM', 'OILPRICEx', 'PPICMM',  'CPIAUCSL', 'CPIAPPSL', 'CPITRNSL', 'CPIMEDSL', 'CUSR0000SAC', 'CUUR0000SAD', 'CUSR0000SAS', 'CPIULFSL', 'CUUR0000SA0L2', 'CUSR0000SA0L5', 'PCEPI', 'DDURRG3M086SBEA', 'DNDGRG3M086SBEA', 'DSERRG3M086SBEA']
t_code_6 = ['CES0600000008', 'CES2000000008', 'CES3000000008', 'M1SL', 'M2SL', 'AMBSL', 'TOTRESNS','BUSLOANS', 'REALLN', 'NONREVSL', 'MZMSL', 'DTCOLNVHFNM', 'DTCTHFNM', 'INVEST']
t_code_7 = ['NONBORRES']
```

```python
# no modification
for col in t_code_1:
    name_col = str(col) + '_st'
    database[name_col] = database[col]

# simple differenciation 
for col in t_code_2:
    name_col = str(col) + '_st'
    database[name_col] = database[col].diff()
    
# double differenciation
for col in t_code_3:
    name_col = str(col) + '_st'
    database['temp'] = database[col].diff()
    database[name_col] = database['temp'].diff()
    database.drop(['temp'], axis = 1)
    
# logarithm
for col in t_code_4:
    name_col = str(col) + '_st'
    database[name_col] = np.log(database[col])
    
# simple differenciation of logarithms
for col in t_code_5:
    name_col = str(col) + '_st'
    database['temp'] = np.log(database[col])
    database[name_col] = database['temp'].diff()
    database.drop(['temp'], axis = 1)
    
# double differenciation of logarithms
for col in t_code_6:
    name_col = str(col) + '_st'
    database['temp1'] = np.log(database[col])
    database['temp2'] = database['temp1'].diff()
    database[name_col] = database['temp2'].diff()
    database.drop(['temp1'], axis = 1)
    database.drop(['temp2'], axis = 1)
    
# simple differenciation of growth rate
for col in t_code_7:
    name_col = str(col) + '_st'
    database['temp'] = database[col].pct_change()
    database[name_col] = database['temp'].diff()
    database.drop(['temp'], axis = 1)
```

# 1. c) principal components

"_In addition, we include as potential predictors the four principal component factors computed from this set of variables._"

```python
X = database.iloc[: , :nb_col] # consider only original variables, not the stationnarized ones
X = X.dropna(axis='columns') # remove columns with empty values
std_slc = StandardScaler()
X_std = std_slc.fit_transform(X) # remove the outliners and scale the data (mean zero, standard deviation one)

print(X_std.shape)

pca = decomposition.PCA(n_components=4) # 4 principal component factors

X_std_pca = pca.fit_transform(X_std) # create the principal components

print(X_std_pca.shape)
print(X_std_pca)
```

```python
# add these four principal components to the database
pc = X_std_pca.transpose()
database = database.assign(pc1=pc[0], pc2=pc[1], pc3=pc[2], pc4=pc[3])
```

"_We consider four lags of all variables, as well as four autoregressive terms. Hence, the analysis contemplates 508 potential predictors._"

i.e. to predict inflation at date t, we use inflation from t-1 to t-5 (4 predictors) as well as all other variables (the 122 original ones + the four principal components) from t-1 to t-5. So in total we have 4+4*(122+4)=508 variables in theory, but we don't have 122 variables (only 113 or 127).

# 1. d) inflation

"_Ï€t is the inflation in month t computed as Ï€t = log(Pt) âˆ’ log(Ptâˆ’1), and Pt is a given price index in period t. The baseline
price index is the CPI._"

```python
# build inflation variable
database['temp'] = np.log(database['CPIAUCSL'])
database['inflation'] = database['temp'].diff()
database.drop(['temp'], axis = 1)
```
